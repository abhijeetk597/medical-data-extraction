```python
from pdf2image import convert_from_path
import pytesseract
import matplotlib.pyplot as plt

POPPLER_PATH = r"C:\poppler-24.02.0\Library\bin"
TESSERACT_ENGINE_PATH = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
pytesseract.pytesseract.tesseract_cmd = TESSERACT_ENGINE_PATH
```

```python
DOC_PATH = r"E:\Projects\medical-data-extraction\backend\resources\patient_details\pd_2.pdf"
pages = convert_from_path(DOC_PATH, poppler_path=POPPLER_PATH)

plt.figure(figsize=(15,10))
plt.imshow(pages[0])
plt.axis(False);
```

```python
text = pytesseract.image_to_string(pages[0], lang="eng")
print(text)
```

```python
import numpy as np
import cv2

def preprocess_image(img):
    gray = cv2.cvtColor(np.array(img), cv2.COLOR_BGR2GRAY)
    resized = cv2.resize(gray, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_LINEAR)
    processed_image = cv2.adaptiveThreshold(resized,
                                           255,
                                           cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                           cv2.THRESH_BINARY,
                                           65,
                                           13)
    return processed_image
```

```python
img = preprocess_image(pages[0])
text = pytesseract.image_to_string(img, lang="eng")
print(text)
```
